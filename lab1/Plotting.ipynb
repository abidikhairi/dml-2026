```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# plot sin and cos functions
```

```python
x = np.linspace(-10, 10, 1000)

x.shape
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.legend()
plt.show()
```

```python
### Analyzing the stock market data

silver_prices_path = '../data/prices/silver_prices_historical.csv'
```

```python
silver_df = pd.read_csv(silver_prices_path, date_format='%Y-%m-%d')

silver_df
```

```python
# silver_df1 = silver_df[silver_df['Date'] <= '2016-01-31']
```

```python
dates = silver_df['Date']
open_prices = silver_df['Open']
close_prices = silver_df['Close']
```

```python
plt.figure(figsize=(10, 6))

plt.plot(dates, open_prices, label='Open Price')
plt.plot(dates, close_prices, label='Close Price')

plt.legend()
plt.show()
```
